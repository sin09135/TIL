#### 정규표현식은 여러가지 특수문자를 이용해서 확장한다.
- 문자열 앞에 ^를 붙이면 문자열이 맨 앞에 오는지 판단
- 문자열 뒤에 $를 붙이면 문자열이 맨 뒤에 오는지 판단

- match : 가장 맨 앞에 인덱스에서 해당하는 패턴이 있는지 보고 없으면 바로 스킵
- search : 전체를 확인하고 매칭이 되는 패턴을 찾는다.

```python
re.search('hi', 'hi,guys')
```

```python
re.match('hi','hello, hi')
```

- 여러 문자열 중 하나라도 포함되면 찾기

```python
re.match('hello|hi','hi,python')
```

```python
re.match('hello|hi','python,hi') # 제일 앞에서 찾고 없으면 아무것도 출력 안함
```

#### 1개 이상의 패턴을 찾아야 하는 경우
- 정규식에서 숫자를 표현할 때 [] 대괄호 안에 값을 넣어서 패턴을 표시한다.
- 숫자, 한글, 알파벳, 특수문자
- 모든 숫자를 표현하는 방법 : [0-9]

- '*':이 특수문자가 0개 이상 있는지 확인
- '+ ': 1개 이상 있는지 확인


```python
re.match('[0-9]','8956') # 맨 앞에만 찾는다. 따라서 8
```

```python
re.search('[0-9]','8956')
```

```python
re.match('[0-9]+','8956') # + : 1 이상 확인
```

```python
re.search('[0-9]+','8956')
```

```python
re.match('[0-9]+','전화번호뒷자리8956') # 맨 앞자리만 비교 > 아무것도 출력되지 않음
```

```python
re.search('[0-9]+','전화번호뒷자리8956')
```

```python
re.search('[가-힣]', '1234안녕hello')
```

```python
re.search('[가-힣]+', '1234안녕hello')
```

```python
re.search('[가-힣]+', '1234안녕hello안녕')
```

```python
re.search('[a-zA-z가-힣]+','1234frameData안녕')
```

### 문자 개수 판단하기

```python
re.search('c{4}','cccc') # c를 다섯번 찾아줘
```

```python
re.search('c{5}','cccb') #c가 5번 없어서 촐력되지 않음
```

```python
# 주민등록번호 패턴
'123456 - 1234567'
re.search('^010-[0-9]{4}-[0-9]{4}$','010-1234-1234')
```

#### not : 특정 문자 범위 포함되지 않는지 판단하는 것
- [^범위]*
- [^범위]+

```python
re.search('[A-Z]+','안녕123')
```

```python
re.search('[^A-Z]+','안녕123') # not 조건
```

- 특수문자 찾기
  - /특수문자
  - 영문, 한글, 숫자 더 간단하게 표현 가능
    - \d [0-9]
    - \D [^0-9]
    - \w [a-Za-zo-a9]
    - \W [^a-Za-zo-a9]]
  

```python
re.search('\d','8956')
```

```python
re.search('\d+','8956') # + 붙이면 전체 다 찾음
```

- 공백처리
  - ''
  - \s
  - \S

```python
re.search('([0-9]+) ([a-z]+)', '1234 kim') #공백 있음
```

```python
test1 = re.search('([0-9]+) ([a-z]+)', '1234kim') #공백 없음
```

```python
test1.group(1)
```

### 정규식을 이용한 문자열 패턴 문법

```python
tests = re.findall('[a-z]+', '1234 a 12345 b 12345 c')
tests
```

#### 문자열 바꿔보기
- 기존 문자열을 내가 원하는 문자열로 바꿀 수 있다.
- 전체 데이터에서 내가 특정하는데이터만 원하는 값으로 바꿀 수 있다.
- 민감한 정보는 다 찾아서 바꾸는 것
- re.sub('패턴','바꿀문자열',바꿀횟수)

```python
re.sub('xxx | yyy','cake','aaa bbb ccc xxx yyyy')
```

```python
re.sub('[a-z]+','영어',' 1 2 3 4 a b')
```
