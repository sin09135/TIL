## 1) 사용자 정의 함수
  - addition()

```python
def addition(m,n):
  k = m+n
  return k
```

```python
addition(m = 3, n = 6)
```

- subtraction()

```python
def subtraction(m,n):
  k = m - n
  return k
```

```python
# 한줄로
subtraction = lambda m,n : m-n
```

```python
subtraction(3,2)
```

```python
subtraction(m = 3, n = 3)
```

- 입력값 개수 미지정

```python
def add_all(*inputs):
  result = 0

  for i in inputs:
    result += i

  return result

```

```python
add_all(1,2,3,4)
```

- 매개변수 초깃값 설정

```python
def add_mul(*inputs, operator = 'addition'):
  if operator == 'addition':
    result = 0
    for i in inputs:
      result = result + i
  elif operator == 'multiplication':
    result = 1
    for i in inputs:
      result = result * i
  return result
```

```python
add_mul(3,5, operator = 'multiplication')
```
