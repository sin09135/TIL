### 반복문 이용, 시각화 그래프

```python
import matplotlib.gridspec as gridspec

grid = gridspec.GridSpec(2,2) # 4가지 그래프
plt.figure(figsize = (16,10))
plt.subplots_adjust(wspace = 0.5, hspace = 0.4) # 그래프 충돌방지

mpg_features = ['manufacturer','model','category','drv'] # 범주형 데이터 선택

# sns countplot 그리기
for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    
    # countplot parameters
    sns.countplot(x = feature,
                  data = df,
                  hue = 'drv',
                  palette = 'pastel',
                  ax = ax)
    plt.xticks(rotation = 90) # x 축 값 축 전환
    ax.set_title(f'{feature} Distribution')
```

### 반복문 사용, 간단한 시각화

#### countplot

```python
cols = ['survived','pclass','sex']

fig, axs = plt.subplots(nrows = 1, ncols = len(cols), figsize = (15,3))

for idx,columns in enumerate(cols):
    print('idx',idx)
    sns.countplot(x = columns, data = df, ax = axs[idx])
```

#### barplot

```python
cols = ['survived','pclass','sex']

fig, axs = plt.subplots(nrows = 1, ncols = len(cols), figsize = (15,3))

for idx,columns in enumerate(cols):
    print('idx',idx)
    sns.barplot(x = columns,y = 'survived', estimator = sum , data = df, ax = axs[idx])
```

## 시계열 데이터
- 시간의 흐름을 보여주는 인덱스, datetime 을 이용해 데이터 확인
- 보간법처럼, 시간의 흐름에 따라 데이터의 특정이 생성됨
    - 데이터를 마음대로 바꿀 시, 시간의 특성이 사라짐
- 예측을 진행할 때, 문제가 될 수 있다(ML train_test_split)
    - 시계열은 랜덤하게 나눌 수가 없다.
    - 시계열은 time과 관련하게 나눠서 데이터를 분할한다.
    
- datetime, TimeStamp, Timedelta 등 여러가지 제공중
    - datetime : data + time(연월일시분초) -> 가장 많이 사용
    - TimeStamp : 날짜와 시간, 단일객체
    - Timedelta : 객체를 가지고 있으면서 시간의 계산이 가능함(덧셈, 뺄셈)
    
### 시계열 분석

### 실제 데이터로 실습

```python
df = pd.read_csv('data/crime.csv')
```

```python
df.head()
```

##### 1. 시계열 데이터로 만들기(datatime)

```python
df.info()
```

```python
df['REPORTED_DATE']= pd.to_datetime(df['REPORTED_DATE'])
```

##### 2. 시계열 데이터 index 지정
- 시계열 데이터를 쉽게 가공하기 위해서는 인덱스 지정이 필요하다.

```python
df = df.set_index('REPORTED_DATE')
```

```python
df
```

#### 3. 인덱스 사용

##### 원하는 시간대 확인(between_time)
