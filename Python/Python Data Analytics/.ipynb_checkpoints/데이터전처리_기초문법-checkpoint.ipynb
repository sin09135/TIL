## 4. 파생변수 
- 판다스에서 컬럼과 컬럼을 가지고 바로 작업 가능
- apply 함수를 이용해서 lambda를 적용
- loc, iloc 사용
- assign() 함수 사용

### 파생변수
- 외부변수, 내부 데이터를 가지고 새로운 변수를 만드는 것

#### 파생변수를 만드는 이유
- 1. 데이터 분석에 필요한 경우 - 모델링 및 피쳐 엔지니어링 등

#### 방법1. 판다스에서 컬럼과 컬럼을 가지고 바로 작업 가능

```python
df['3_mean'] = (df['math'] + df['english'] + df['science']) / 3 # 시리즈 형태로 작업
```

```python
df
```

#### 방법2. assign 함수 이용

```python
df['3_mean'].mean()
```

```python
df = df.assign(mean_t = (df['math'] + df['english'] + df['science']) / 3, inplace = True, no = 1)
```

```python
df
```

```python
import numpy as np
```

- 평균을 기준으로 평균 이상, 평균 이하

```python
df.assign(test = np.where(df['3_mean'] > 67.3, '평균이상','평균이하'))
```

```python
np.where(df['3_mean'] >= 67.3)
# 넘파이 함수를 이용해서 index로 위치를 잡고 그 잡은 위치에 값을 넣는 개념
```

#### 방법 3. lambda

```python
df.assign(test1 = lambda x : (x['math'] + x['english'] + x['science'])/3)
```

#### 방법 4. groupby
- 그룹으로 묶음 > 묶어서 계산할 수 있음
- 기준이 필요
- agg 함수와 같이 사용하는 경우 많다.
- df.groupby('기준').agg(

```python
df.groupby('nclass')# nclass 를 묶으면 반별 결과가 나옴
```

```python
df.groupby('nclass').agg(ncl_math = ('math','mean'))
```

```python
df.groupby('nclass').agg(ncl_math_std = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```
