## titanic_visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

### 범주형 변수 시각화

```python
import matplotlib.gridspec as gridspec

grid = gridspec.GridSpec(2,2) # 행,열 개수 > 4가지 그래프

# 도화지 그리기
plt.figure(figsize = (16,10))
plt.subplots_adjust(wspace = 0.5, hspace = 0.4)

# 데이터 선택
titanic_features = ['survived','sex','pclass','embarked']

# seaborn countplot 그리기
for idx, feature in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    # countplot parameters
    sns.countplot(x = feature,
                  data = df,
                  hue = 'survived',
                  palette = 'pastel',
                  ax = ax)
    # plt.xticks(rotation = 90) - x축 글자 회전
    ax.set_title(f'{feature} Distribution')
```
