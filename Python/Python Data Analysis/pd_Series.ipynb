```python
import pandas as pd
import numpy as np

df = pd.read_csv('weather.csv')
```

```python
df.head()
```

```python
combined_df = df.append(df, ignore_index=True)
print("\n데이터프레임 연결:\n", combined_df)

# - describe
print("\n기술 통계량 요약:\n", df.describe())

# - drop_duplicates (중복 제거)
unique_df = df.drop_duplicates()
print("\n중복 제거한 데이터프레임:\n", unique_df)

# - get_values (시리즈의 값 구하기)
values_array = df['d1'].values
print("\n시리즈의 값 얻기:\n", values_array)

# - sort_values
sorted_df = df.sort_values(by='d2', ascending=True)
print("\n값 정렬:\n", sorted_df)

# - to_frame
df_as_frame = df['d1'].to_frame()
print("\n시리즈를 DataFrame으로 변환:\n", df_as_frame)
```
