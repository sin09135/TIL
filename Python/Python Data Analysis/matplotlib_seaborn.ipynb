#### matplotlib inline

```python
import matplotlib.pyplot as plt
```

```python
# 큰 그림부터 세부적으로 만지는 느낌
plt.scatter([1,2,3,4],[5,6,7,8])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### 축의 범위 설정!
- x축과 y축이 내가 원하는 범위의 값만 나타낼 수 있다
- 축의 범위를 조정 가능하다.
- xlim, ylim

```python
plt.scatter([1,2,3,4],[5,6,7,8])
plt.xlim(0,3)
plt.ylim(5,7)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```
