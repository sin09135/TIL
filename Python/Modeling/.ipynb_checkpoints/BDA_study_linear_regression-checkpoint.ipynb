1. 선형회귀의 가정에 대해 쓰고, 그 가정들을 설명해주세요.

- (1) 선형성(2) 독립성(3) 등분산성(4) 정규성

- 선형성(Linear Relationship): 종속 변수와 독립 변수 간의 관계가 선형 > 변수들 간의 관계를 직선으로 표현 가능해야 함
- 독립 변수들 간의 다중공선성(Multicollinearity): 독립 변수들 간에 강한 상관관계가 없어햐 함 > 회귀계수의 추정치를 불안정
- 잔차의 정규성(Normality of Residuals):잔차가 정규분포
- 잔차의 등분산성(Homoscedasticity):잔차의 분산이 일정 > 모든 수준의 독립 변수에 대해 오차의 분산이 일정
- 잔차의 독립성(Independence of Errors):잔차들 간에 자기 상관이 없어야 함 > 한 관측치의 오차가 다른 관측치의 오차와 독립적
- 독립 변수의 독립성(Independence of Predictors):독립 변수들 간에 강한 상관관계가 없다는 가정

2. 더빈-왓슨 테스트가 무엇인지, 무엇을 위해 쓰는지 설명해주세요.

- 회귀 분석에서 잔차의 자기 상관(autocorrelation)을 검정하기 위해 사용되는 테스트
- 자기 상관성은 시간적 순서가 있는 데이터에서 연속된 관측값들 사이의 상관관계를 의미

3.  statsmodels를 이용하여 선형회귀를 진행하고, 모델의 가정을 검증 해주세요.(scatter, shapiro-willk, 더빈 왓슨, q-q, 잔차의 분산)

```python
import numpy as np #모듈불러오기
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from scipy.stats import shapiro
from statsmodels.stats.stattools import durbin_watson
from scipy.stats import shapiro, probplot
```

```python
california = fetch_california_housing()
data = pd.DataFrame(california.data, columns=california.feature_names)
data['MedHouseVal'] = california.target
```

```python
x = data['MedInc']
y = data['MedHouseVal']
```

```python
# 상수항 추가
X = sm.add_constant(x)

# 선형 회귀 모델 피팅
model = sm.OLS(y, X).fit()

# 모델 요약 출력
print(model.summary())
```

### 1. 선형성 (Linearity) - 산점도, 회귀선

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['MedInc'], y=data['MedHouseVal'], alpha=0.6)
plt.plot(data['MedInc'], model.fittedvalues, color='red', linewidth=2)
plt.title('Scatter plot with Regression Line')
plt.xlabel('MedInc')
plt.ylabel('MedHouseVal')
plt.show()
```
