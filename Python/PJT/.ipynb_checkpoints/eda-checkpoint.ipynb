```python
import pandas as pd

# 각 데이터프레임에 적용할 컬럼 이름 변경 사전
columns_apple = {'평균': '사과_가격', '전통시장': '사과_전통시장', '유통업체': '사과_유통업체'}
columns_pear = {'평균': '배_가격', '전통시장': '배_전통시장', '유통업체': '배_유통업체'}
columns_gam = {'평균': '단감_가격', '전통시장': '단감_전통시장', '유통업체': '단감_유통업체'}
columns_shine = {'평균': '포도_가격', '전통시장': '포도_전통시장', '유통업체': '포도_유통업체'}
columns_orange = {'평균': '오렌지_가격', '전통시장': '오렌지_전통시장', '유통업체': '오렌지_유통업체'}
columns_pineapple = {'평균': '파인애플_가격', '전통시장': '파인애플_전통시장', '유통업체': '파인애플_유통업체'}
columns_mango = {'평균': '망고_가격', '전통시장': '망고_전통시장', '유통업체': '망고_유통업체'}

# 데이터프레임 읽기
apple = pd.read_excel('data/사과_가격.xlsx')
pear = pd.read_excel('data/배_가격.xlsx')
gam = pd.read_excel('data/단감.xlsx')
shine = pd.read_excel('data/포도_가격.xlsx')
orange = pd.read_excel('data/오렌지.xlsx')
pineapple = pd.read_excel('data/파인애플.xlsx')
mango = pd.read_excel('data/망고.xlsx')

# 컬럼 이름 변경
apple.rename(columns=columns_apple, inplace=True)
pear.rename(columns=columns_pear, inplace=True)
gam.rename(columns=columns_gam, inplace=True)
shine.rename(columns=columns_shine, inplace=True)
orange.rename(columns=columns_orange, inplace=True)
pineapple.rename(columns=columns_pineapple, inplace=True)
mango.rename(columns=columns_mango, inplace=True)

# 데이터프레임 병합
merged_df = apple.merge(pear, on='구분 ', how='outer')
merged_df = merged_df.merge(gam, on='구분 ', how='outer')
merged_df = merged_df.merge(shine, on='구분 ', how='outer')
merged_df = merged_df.merge(orange, on='구분 ', how='outer')
merged_df = merged_df.merge(pineapple, on='구분 ', how='outer')
merged_df = merged_df.merge(mango, on='구분 ', how='outer')

# 결과 확인
merged_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 각 데이터프레임에 적용할 컬럼 이름 변경 사전
columns_apple = {'평균': '사과_가격', '전통시장': '사과_전통시장', '유통업체': '사과_유통업체'}
columns_pear = {'평균': '배_가격', '전통시장': '배_전통시장', '유통업체': '배_유통업체'}
columns_gam = {'평균': '단감_가격', '전통시장': '단감_전통시장', '유통업체': '단감_유통업체'}
columns_shine = {'평균': '포도_가격', '전통시장': '포도_전통시장', '유통업체': '포도_유통업체'}
columns_orange = {'평균': '오렌지_가격', '전통시장': '오렌지_전통시장', '유통업체': '오렌지_유통업체'}
columns_pineapple = {'평균': '파인애플_가격', '전통시장': '파인애플_전통시장', '유통업체': '파인애플_유통업체'}
columns_mango = {'평균': '망고_가격', '전통시장': '망고_전통시장', '유통업체': '망고_유통업체'}

# 데이터프레임 읽기
apple = pd.read_excel('data/사과_가격.xlsx')
pear = pd.read_excel('data/배_가격.xlsx')
gam = pd.read_excel('data/단감.xlsx')
shine = pd.read_excel('data/포도_가격.xlsx')
orange = pd.read_excel('data/오렌지.xlsx')
pineapple = pd.read_excel('data/파인애플.xlsx')
mango = pd.read_excel('data/망고.xlsx')

# 컬럼 이름 변경
apple.rename(columns=columns_apple, inplace=True)
pear.rename(columns=columns_pear, inplace=True)
gam.rename(columns=columns_gam, inplace=True)
shine.rename(columns=columns_shine, inplace=True)
orange.rename(columns=columns_orange, inplace=True)
pineapple.rename(columns=columns_pineapple, inplace=True)
mango.rename(columns=columns_mango, inplace=True)

# 데이터프레임 병합
merged_df = apple.merge(pear, on='구분 ', how='outer')
merged_df = merged_df.merge(gam, on='구분 ', how='outer')
merged_df = merged_df.merge(shine, on='구분 ', how='outer')
merged_df = merged_df.merge(orange, on='구분 ', how='outer')
merged_df = merged_df.merge(pineapple, on='구분 ', how='outer')
merged_df = merged_df.merge(mango, on='구분 ', how='outer')

# '구분' 열을 제외한 모든 열에서 쉼표 제거 후 숫자로 변환
for col in merged_df.columns:
    if col != '구분 ':
        merged_df[col] = merged_df[col].str.replace(',', '').astype(float)

# "가격"이 포함된 열만 선택
price_cols = [col for col in merged_df.columns if '가격' in col]

# 선택한 열들로 상관관계 계산
correlation_matrix = merged_df[price_cols].corr()

# 히트맵 그리기
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title('과일 가격 상관관계 히트맵')
plt.show()
```
