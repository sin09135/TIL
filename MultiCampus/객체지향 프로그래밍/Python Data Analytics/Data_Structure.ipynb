> ## 5) Formatting

* '%s': String

```
'The X-Files is an American science fiction %s.' % 'Drama'
```

# IV. Dictionary

* <font color='red'>순서가 없음(Unordered)</font>
* <font color = 'blue'>Key와 Value 한쌍으로 구성

> ## 1) {Key:Value} 구조 선언

```
D1 = {'Name':'LEE', 'Age':24}
print(D1)
```

```
print(type(D1))
print(type(D1['Name']))
print(type(D1['Age']))
```

```
D1['Name']
```

> ## 2) Key:Value 추가

```
D1['Height'] = 183

print(D1)
```

> ## 3) Key:Value 삭제

```
del D1['Age']
print(D1)
```

> ## 4) Function( )

* Key 확인

```
D1.keys()
```

* Value 확인

```
D1.values()
```

* Key:Value 삭제

```
print(D1)
D1.clear()
print(D1)
```

> ## 5) Dictionary with List

```
L1 = ['Red', 'Green', 'Blue']
L2 = [255, 127, 63]
```

```
D2 = {x : y for x, y in zip(L1, L2)}
print(D2)
```

# V. Casting

> ## 1) Data Type

* int to float

```
print(type(9))
print(float(9))
```

* str to float

```
print(type('9.4'))
print(float('9.4'))
```

* float to int

```
print(type(9.0))
print(9.0)
print(type(int(9.0)))
print(int(9.0))
```

* str to int

```
print(type('9'))
print('9')
print(type(int('9')))
print(int('9'))
```

* <font color='blue'>float to int</font>
* <font color = 'red'>Warning!!!

```
print(type(3.14))
print(3.14)
print(type(int(3.14)))
print(int(3.14))
```

* int to str

```
print(type(9))
print(type(str(9)))
```

* float to str

```
print(type(3.14))
print(type(str(3.14)))
```
