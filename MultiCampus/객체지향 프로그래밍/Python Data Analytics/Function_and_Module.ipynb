# I. def

* 함수(Function): 지정된 <font color='red'>입력값</font>을 받아 처리 후 <font color='blue'>출력값</font> 리턴
 - <font color='red'>입력값</font> -> 함수(Function) -> <font color='blue'>출력값</font>

> ## 1) 사용자 정의 함수 with <font color='blue'>def

* addition( ) 정의 및 사용

```
def addition(m, n):
  k = m + n
  return k
```

```
addition(m = 3, n = 6)
```

> ## 2) 사용자 정의 함수 with <font color='red'>lambda

* subtraction( ) 정의 및 사용

```
# def subtraction(m, n):
#   k = m - n
#   return k
```

* <font color='red'>lambda</font>: 한 줄로 함수 정의

```
subtraction = lambda m, n : m - n
```

> ## 3) 매개변수 위치

* m - n

```
subtraction(m = 3, n = 6)
```

* m - n

```
subtraction(n = 6, m = 3)
```

* m = 3, n = 6

```
subtraction(3, 6)
```

* <font color = 'red'>m = 6, n = 3

```
subtraction(6, 3)
```

> ## 4) 입력값 개수 미지정

* <font color = 'red'>'*' 기호</font>가 입력값을 모아서 <font color='blue'>Tuple로 생성

```
def add_all(*inputs):
    result = 0

    for i in inputs:
        result = result + i
    return result
```

```
add_all(1, 3, 5, 7, 9)
```

> ## 5) 매개변수 <font color = 'blue'>초깃값</font> 설정

* <font color = 'blue'>operator = 'addition'

```
def add_mul(*inputs, operator = 'addition'):
    if operator == 'addition':
        result = 0
        for i in inputs:
            result = result + i
    elif operator == 'multiplication':
        result = 1
        for i in inputs:
            result = result * i
    return result
```

* <font color = 'blue'>마지막 매개변수 생략

```
add_mul(1, 3, 5, 7, 9)
```

* 초깃값이 아닌경우 매개변수 지정

```
add_mul(1, 3, 5, 7, 9, operator = 'multiplication')
```

* 초깃값 설정 - m = 3, n = 6

```
def addition(m = 3, n = 6):
  k = m + n
  return k
```

* <font color='red'>매개변수 없이 실행

```
addition()
```

* <font color='blue'>매개변수 지정 실행

```
addition(m = 7, n = 9)
```

# II. Module
* <font color='red'>Module</font>: 사용자 정의 함수를 포함하고 있는 파이썬 스크립트(*.py)
* <font color='blue'>import</font>: 파이썬 스크립트에 정의된 함수를 메모리로 호출

> ## 1) 내장 모듈

* 사용구문: <font color = 'blue'>import</font> Module_Name

```
import time
```

* 사용구문: Module_Name<font color = 'red'>.</font>Function_Name( )

```
print('스크립트 시작')
time.sleep(9)
print('시작 9초 후 스크립트 종료')
```

* Built-in Module Path

```
import sys

sys.path
```

> ## 2) 내장 모듈 with Alias (<font color='blue'>import ~ as</font>)

* 사용구문: <font color = 'blue'>import</font> Module_Name <font color = 'blue'>as</font> Alias

```
import time as t
```

* 사용구문 : <font color = 'blue'>Alias</font><font color = 'red'>.</font>Function_Name( )

```
print('5초 후에 결과 출력')
t.sleep(5)
print('축하합니다. 합격입니다!')
```
