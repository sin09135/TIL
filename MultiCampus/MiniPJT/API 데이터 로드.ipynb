## 유동인구 센서 데이터 로드(07.03 ~ 07.30, 4주간)

```python
sdot1 = pd.read_csv('/Users/kimsinwoo/Desktop/multicampus/Project/Mini PJT/raw data/S-DoT_WALK_2023.07.03-07.09.csv', encoding = 'euc-kr')
sdot2 = pd.read_csv('/Users/kimsinwoo/Desktop/multicampus/Project/Mini PJT/raw data/S-DoT_WALK_2023.07.10-07.16.csv', encoding = 'euc-kr')
sdot3 = pd.read_csv('/Users/kimsinwoo/Desktop/multicampus/Project/Mini PJT/raw data/S-DoT_WALK_2023.07.17-07.23.csv', encoding = 'euc-kr')
sdot4 = pd.read_csv('/Users/kimsinwoo/Desktop/multicampus/Project/Mini PJT/raw data/S-DoT_WALK_2023.07.24-07.30.csv', encoding = 'euc-kr')
```

```python
data_frames = [sdot1, sdot2, sdot3, sdot4]
sdot_4w_df = pd.concat(data_frames, ignore_index=True)
```

```python
sdot_4w_df
```

```python
sdot_4w_df.info() #결측치 없음, 총 데이터 368,816개
```

### 측정시간 나누기

```python
sdot_4w_df['날짜'] = sdot_4w_df['측정시간'].str.split('_').str[0]
```

```python
sdot_4w_df['일자']= sdot_4w_df['날짜'].str.split('-').str[2]
```

```python
sdot_4w_df.head()
```

```python
sdot_4w_df['날짜'].value_counts().sort_values()
```

#### 7월 2일 측정값이 7월 3일로 넘어간 것 같다고 예상, 그리고 15일 측정값이 없음 확인 Task

```python
# 시간별 시각화
fig, ax = plt.subplots(figsize = (20,8))
sns.barplot(x = '일자',y = '방문자수',data = sdot_4w_df,ax = ax)
plt.show()
```

```python
# 지역별 시각화 - traditional market, public_facilities
fig, ax = plt.subplots(figsize = (10,8))
sns.barplot(x = '지역',y = '방문자수',data = sdot_4w_df,ax = ax)
plt.show()
```

```python
sdot_4w_df['자치구'].value_counts()
```

```python
# 자치구별 시각화 - 송파구, 동작구 두드러짐
fig, ax = plt.subplots(figsize = (20,8))
sns.barplot(x = '자치구',y = '방문자수',data = sdot_4w_df,ax = ax)
plt.show()
```

```python
# 행정동 별 시각화
fig, ax = plt.subplots(figsize = (20,8))
sns.barplot(x = '행정동',y = '방문자수',data = sdot_4w_df,ax = ax)
plt.show()
```

```python
# 행정동별 방문자수
sdot_행정동 = sdot_4w_df.groupby('행정동')['방문자수'].sum().sort_values(ascending=False)
```

```python
sdot_행정동 = pd.DataFrame(sdot_4w_df.groupby('행정동')['방문자수'].sum().sort_values(ascending=False))
```
