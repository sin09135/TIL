## 서울시 상권분석 서비스 추정 유동인구 API

```python
import requests 
import pandas as pd
import os

# json 파일, 1 ~ 1000 개의 데이터를 가져옴
data_dict = {}

for i in range(1, 410): # 약 40만개 데이터 추출
    end_num = i * 1000
    start_num = end_num - 999
    url = f'http://openapi.seoul.go.kr:8088/7949646757736b773733414e71744d/json/IotVdata018/{start_num}/{end_num}'
    res = requests.get(url) 
    sdot_data = res.json()
    data_dict[start_num] = pd.DataFrame(sdot_data['IotVdata018']['row'])

# Concatenate all DataFrames in the dictionary and reset index
sdot_df = list(data_dict.values())
sdot_df_sample = pd.concat(sdot_df, ignore_index=True)

# Export the merged DataFrame to a CSV file
sdot_df_sample.to_csv('/Users/kimsinwoo/Desktop/multicampus/Project/Mini PJT/sdot_df_sample.csv', index=False)

sdot_df_sample.head(5)
```
